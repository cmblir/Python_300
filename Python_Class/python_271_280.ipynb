```python
# 280
import random
import math

lst = []

class Account:
    account_count = 0
    interest = 0
    interest_sum = 0
    deposit_count = 0
    withdraw_count = 0
    deposit_lst = []
    withdraw_lst = []

    def __init__(self, name, balance):
        self.name = str(name)
        self.balance = balance
        self.bank = "농협은행"
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)

        self.account_num = '계좌번호: ' + num1 + '-' + num2 + '-' + num3
        Account.account_count += 1

    @classmethod  # 클래스 선언
    def get_account_num(cls):
        return cls.account_count

    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            Account.interest += 1
            Account.deposit_count += 1
            Account.deposit_lst.append(str(Account.deposit_count) + '번째: ' + str(amount) + '원이 입금된 기록이 있습니다.')
        if Account.interest >= 5:
            self.balance *= 1.01
            Account.interest_sum = (self.balance - (self.balance / 1.01))

    def withdraw(self, amount):
        if self.balance > amount:
            self.balance -= amount
            Account.withdraw_count += 1
            Account.withdraw_lst.append(str(Account.withdraw_count) + '번째: ' + str(amount) + '원이 출금된 기록이 있습니다.')
        elif self.balance - amount == 0:
            self.balance -= amount
            print("모든 잔고를 출금했습니다.")
        else:
            print("잔액이 부족합니다.")

    def display_info(self):
        print(f"은행이름: {self.bank}\n예금주: {self.name}\n{self.account_num}\n잔고: {format(math.floor(self.balance), ',')}원")

    @classmethod
    def deposit_history(cls):
        for amount in cls.deposit_lst:
            print(amount)

    @classmethod
    def withdraw_history(cls):
        for amount in cls.withdraw_lst:
            print(amount)

    @classmethod
    def interest_history(cls):
        print(f"{str(math.floor(cls.interest_sum))}원")

    def total(self):
        print("입금된 기록")
        Account.deposit_history()
        print("출금된 기록")
        Account.withdraw_history()
        print("총 입금된 이자")
        Account.interest_history()
        print("현재 고객님의 계좌상태")
        Account.display_info(self)

user = Account("유시녕", 0)
user.deposit(24000)
user.withdraw(10000)
user.deposit(36000)
user.withdraw(20000)
user.deposit(30000)
user.withdraw(30000)
user.deposit(30000)
user.withdraw(40000)
user.deposit(24000)
user.withdraw(4000)
user.total()
```
